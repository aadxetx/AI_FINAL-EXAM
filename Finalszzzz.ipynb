```python
############ QUESTION - 4 ############

import numpy as np


class MarkovC(object):
    def __init__(self, transition_matrix, states):
        
        self.transition_matrix = np.atleast_2d(transition_matrix)
        self.states = states
        self.index_dict = {self.states[index]: index for index in range(len(self.states))}
        
        self.state_dict = {index: self.states[index] for index in range(len(self.states))}
        
    def next_s(self, Curr_states):
        return np.random.choice(self.states,
                               p = self.transition_matrix[self.index_dict[Curr_states], :])
    
    def generate_s(self, Curr_states, n = 10):
        future = []
        for i in range(n):
            next_s = self.next_s(Curr_states)
            future.append(next_s)
            Curr_states = next_s
        return future
    

transition_matrix = [[0.1, 0.7, 0.2],
                     [0.75, 0.15, 0.1],
                     [0.6, 0.35, 0.05]]
city_chain = MarkovC(transition_matrix = transition_matrix, states = ['London', 'Barcelona', 'New York'])

initial_transition = [[0.1, 0.6, 0.3]]
```
